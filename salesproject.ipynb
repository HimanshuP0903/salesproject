```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali_Sales_Data.csv', encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.columns
```

```python
#Gender
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = True)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
#Age Group w.r.t gender
sns.countplot(x = 'Age Group', hue = 'Gender', data = df)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = True)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

```python
#State w.r.t Orders
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(16)
sns.set(rc={'figure.figsize':(26,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
# state w.r.t Amount 
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(16)
sns.set(rc={'figure.figsize':(26,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

```python
# marital status
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(16)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_ms, hue = 'Gender')
```

```python
ax = sns.countplot(x = 'Occupation', data = df)
sns.set(rc={'figure.figsize':(24,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(24,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_occ)
```

```python
sns.set(rc={'figure.figsize':(24,5)})
ax = sns.countplot(x = 'Product_Category', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_occ)
```

```python
sales_occ = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID', y = 'Amount', data = sales_occ)
```
